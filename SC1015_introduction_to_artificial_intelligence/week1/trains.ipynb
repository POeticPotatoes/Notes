```python
import numpy
import pandas as pd
```

```python
url = 'http://en.wikipedia.org/wiki/{}_Summer_Olympics_medal_table'
data = []
for y in range(2000, 2020, 4):
    html_data = pd.read_html(url.format(y))
    medal_data = html_data[2]
    data += [medal_data[:20]]

print("Extracted:",len(data))
for x in data:
    display(x)
```
