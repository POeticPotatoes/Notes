```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import scipy.stats as stats
sb.set()
```

```python
train = pd.read_csv("train.csv", header=0)
```

```python
# Function to fit a regression model on a predictor
def checkRegression(tr, var):
    y = pd.DataFrame(tr[var])
    sp = pd.DataFrame(tr["SalePrice"])
    y_train, y_test, sp_train, sp_test= train_test_split(y, sp, test_size=0.2)
    lr = LinearRegression()  # create object for the class
    lr.fit(y_train, sp_train)  # perform linear regression
    yp = lr.predict(y_test)
    print(var)
    print(lr.coef_)
    print(lr.intercept_)
    plt.scatter(y_test, sp_test)
    plt.plot(y_test, yp, color='red')
    plt.show()
    print("Goodness of Fit of Model \tTest Dataset")
    print("Explained Variance (R^2) \t:", lr.score(y_train, sp_train))
    print("Mean Squared Error (MSE) \t:", mean_squared_error(sp_test, yp))
    print()
```

# Problem 1

```python
checkRegression(train, "GrLivArea")
```

# Problem 2
* TotalBsmtSF is the most suitable predictor among the 3 as it has the lowest Mean Squared Error

```python
checkRegression(train, "TotalBsmtSF")
checkRegression(train, "GarageArea")
```

# Problem 3
* The MSE is drastically improved after removing outliers. This means the model is much better than the one created in Problem 2.

```python
# Remove Outliers
cols = ['SalePrice', 'GrLivArea'] # source: https://stackoverflow.com/questions/35827863/remove-outliers-in-pandas-dataframe-using-percentiles
                                  # (I understand the code though)
Q1 = train[cols].quantile(0.25)
Q3 = train[cols].quantile(0.75)
IQR = Q3 - Q1

ctrain = train[~((train[cols] < (Q1 - 1.5 * IQR)) |(train[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
checkRegression(ctrain, "GrLivArea")
```
